```python
# Define the size of your train and test sets
train_size = int(0.75 * len(non_promo_sales))  # 80% for training
test_size = len(non_promo_sales) - train_size  # 20% for testing
#non_promo_sales['day'] = pd.to_datetime(non_promo_sales['day'])
# Split the data
X_train = non_promo_sales['day'][:train_size]
y_train = non_promo_sales['num_sales'][:train_size]
X_test = non_promo_sales['day'][train_size:]
y_test = non_promo_sales['num_sales'][train_size:]
```

```python
plt.figure(figsize=(10,10))
f, ax = plt.subplots(1, 1)
ax.plot(X_test,y_test,label='Test', color='r')
ax.plot(X_train,y_train,label='Train', color='orange')
plt.ylabel('Num Sales')
ax.legend()
plt.show()
```

```python
import sklearn
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt


lm = linear_model.LinearRegression()
model = lm.fit(X_train.values.reshape(-1, 1),y_train)

predictions = lm.predict(X_test.values.astype(float).reshape(-1, 1))
```

```python
plt.figure(figsize=(10,10))
f, ax = plt.subplots(1, 1)
ax.plot(X_test,predictions,label='Linear fit', lw=3)
ax.plot(X_train,y_train,label='Train sales', color='r')
ax.plot(X_test,y_test,label='Test sales', color='orange')
plt.ylabel('Num Sales')
plt.xlabel('days')
plt.title('Linear Regression Model')
ax.legend()
#score = accuracy_score(y_test,prediction)
#print('R^2 score for perfect model is', score)
plt.show()
```
